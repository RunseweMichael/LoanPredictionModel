```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
loan_data = pd.read_csv("loan_approval_dataset.csv")
loan_data
```

```python
loan_data.columns = loan_data.columns.str.strip()

loan_data = loan_data.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
```

```python
loan_data.isnull().sum()
```

```python
loan_data.describe()
```

```python
loan_data.head()
```

```python
loan_data.info()
```

```python
loan_data["education_binary"] = loan_data["education"].map({"Not Graduate": 0, "Graduate": 1})
```

```python
loan_data.head()
```

```python
loan_data["self_employed_binary"] = loan_data["self_employed"].map({"No": 0, "Yes": 1})
```

```python
loan_data
```

```python
X = loan_data.drop(["loan_id","no_of_dependents","education","self_employed","residential_assets_value","loan_status","education_binary","self_employed_binary","loan_status_binary"], axis=1)
```

```python
X.describe()
```

```python
X.head()
```

```python
loan_data["loan_status_binary"] = loan_data["loan_status"].map({"Rejected": 0, "Approved": 1})
```

```python
loan_data
```

```python
y = loan_data["loan_status_binary"]
```

```python
y.describe()
```

```python
y.head()
```

# Training with Logistic Regression

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform((X_train))
X_test= sc.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
y_pred
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
result = accuracy_score(y_test,y_pred)*100
result
```

```python
cmatrix = confusion_matrix(y_test,y_pred)
cmatrix
```

```python
creport = classification_report(y_test,y_pred)
print(creport)
```

# Visualizing Confusion Matrix

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report


y_pred = lr.predict(X_test)


cm = confusion_matrix(y_test, y_pred)


plt.figure(figsize=(10,7))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=set(y_test), yticklabels=set(y_test))
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()

```

# Classification Report Heat Map

```python
from sklearn.metrics import classification_report
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


report = classification_report(y_test, y_pred, output_dict=True)
df_report = pd.DataFrame(report).transpose()

plt.figure(figsize=(8,6))
sns.heatmap(df_report.iloc[:-1, :-1], annot=True, cmap="Blues")
plt.title("Classification Report Heatmap")
plt.show()
```
